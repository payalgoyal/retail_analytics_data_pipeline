# Data Understanding – Key Observations

## 1. Dataset Overview

The dataset contains **1,067,371 rows and 8 columns**, representing transactional retail sales data.

Each row represents a **product line within an invoice**, meaning a single invoice can contain multiple products purchased by a customer.

The columns present in the dataset are:

* Invoice – Invoice number for each transaction
* StockCode – Unique product identifier
* Description – Product name or description
* Quantity – Number of items purchased
* InvoiceDate – Date and time of the transaction
* Price – Unit price of the product
* Customer ID – Unique customer identifier
* Country – Country where the customer is located

---

## 2. Data Types

The dataset contains three types of data fields:

* **Categorical/Text:** Invoice, StockCode, Description, Country
* **Numerical:** Quantity, Price
* **Customer Identifier:** Customer ID

One important issue identified is that **InvoiceDate is stored as an object instead of a datetime datatype**, which will need to be converted during the data cleaning stage for proper time-based analysis.

---

## 3. Missing Values

Two columns contain missing values:

| Column      | Missing Records |
| ----------- | --------------- |
| Description | 4,382           |
| Customer ID | 243,007         |

Observations:

* Missing values in **Description** are relatively small compared to the dataset size.
* A significant number of records have **missing Customer IDs (~22%)**, which suggests that some purchases were made by **guest customers or unidentified buyers**.

For customer-level analysis (such as repeat purchase behaviour or RFM analysis), rows with missing Customer IDs may need to be excluded.

---

## 4. Duplicate Records

The dataset contains **34,335 duplicate rows**.

Possible reasons include:

* repeated invoice entries
* system duplication
* multiple identical product entries recorded more than once

These duplicates will need to be **investigated and removed during data cleaning** to ensure accurate analysis.

---

## 5. Customer Distribution

The dataset contains **5,942 unique customers**.

This indicates that customer-level analysis such as:

* repeat purchase behaviour
* customer lifetime value
* customer segmentation

can be performed.

---

## 6. Geographic Coverage

The dataset contains customers from **43 different countries**, indicating that the retailer operates internationally.

However, the majority of transactions are likely concentrated in a few primary markets (commonly the United Kingdom in this dataset).

---

## 7. Product Returns and Cancellations

Several records contain **negative quantities**.

Example observation from the dataset:

* Negative quantities appear in invoices that begin with the letter **"C"**.

This indicates that these transactions represent **order cancellations or product returns**.

Thus:

* **Invoice numbers starting with "C" represent cancelled orders**
* Negative quantities correspond to returned items.

These transactions must be handled carefully during analysis to avoid incorrect revenue calculations.

---

## 8. Zero Price Transactions

Some transactions contain **Price = 0**.

These may represent:

* promotional items
* free samples
* system adjustments
* data entry anomalies

These records will need to be examined during the cleaning stage before revenue calculations are performed.

---

## 9. Date Range

Date ranging from start date of Invoice to end date of Invoice in the dataset

## 10. Dataset Characteristics

Based on the initial exploration, the dataset represents a **large-scale retail transactional system** with the following characteristics:

* Line-level transaction records
* Multi-product invoices
* International customer base
* Product returns and cancellations
* Potential duplicate records

This makes the dataset suitable for building a **Retail Intelligence and Sales Analytics dashboard**.

## 11. Data Issues Identified for Cleaning

The following data quality issues were identified and will be addressed in the next stage:
