## Data Modelling

```python
import pandas as pd

file_path = r"D:\Analyst\DigitalProduct\Automated_Retail_Intelligence\data\processed\cleaned_online_data.csv"
df = pd.read_csv(file_path,parse_dates=["InvoiceDate"])
```

```python
date_range = pd.date_range(
    start = df["InvoiceDate"].min(),
    end = df["InvoiceDate"].max(),
    freq = "D"
)

dim_date = pd.DataFrame({"Date": date_range})
```

```python
dim_date["Year"] = dim_date["Date"].dt.year
dim_date["Month"] = dim_date["Date"].dt.month
dim_date["MonthName"] = dim_date["Date"].dt.month_name()
dim_date["Quarter"] = dim_date["Date"].dt.quarter
dim_date["Day"] = dim_date["Date"].dt.day
dim_date["Weekday"] = dim_date["Date"].dt.day_name()
```

```python
dim_product = df[["StockCode", "Description"]].drop_duplicates().copy()
```

```python
dim_customer = df[["Customer ID", "Country"]].drop_duplicates().copy()
```

```python
fact_sales = df[[
    "Invoice",
    "StockCode",
    "Customer ID",
    "InvoiceDate",
    "Quantity",
    "Price",
    "Revenue",
    "IsCancelled"
]].copy()
```

```python
fact_sales["Date"] = fact_sales["InvoiceDate"].dt.date
```

```python
processed_folder = r"D:\Analyst\DigitalProduct\Automated_Retail_Intelligence\data\processed"
dim_date.to_csv(r"D:\Analyst\DigitalProduct\Automated_Retail_Intelligence\data\processed\dim_date.csv", index=False)
dim_product.to_csv(r"D:\Analyst\DigitalProduct\Automated_Retail_Intelligence\data\processed\dim_product.csv", index=False)
dim_customer.to_csv(r"D:\Analyst\DigitalProduct\Automated_Retail_Intelligence\data\processed\dim_customer.csv", index=False)
fact_sales.to_csv(r"D:\Analyst\DigitalProduct\Automated_Retail_Intelligence\data\processed\fact_sales.csv", index=False)
```
