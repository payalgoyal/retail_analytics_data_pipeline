## Data Cleaning

```python
import pandas as pd
import numpy as np
```

```python
file_path = r"D:\Analyst\DigitalProduct\Automated_Retail_Intelligence\data\raw\online_retail.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
df.shape
```

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
```

```python
df["InvoiceDate"].min(), df["InvoiceDate"].max()
```

```python
df["InvoiceDay"] = df["InvoiceDate"].dt.date
```

```python
df["Customer ID"] = df["Customer ID"].astype("Int64").astype("string")
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df = df[df["Description"].notna()]
```

```python
df = df[df["Customer ID"].notna()]
```

```python
df["IsCancelled"] = df["Invoice"].str.startswith("C")
```

```python
df["IsCancelled"].value_counts()
```

```python
df[df["Price"] <= 0].shape
```

```python
df = df[df["Price"] > 0]
```

```python
df["Revenue"] = df["Quantity"] * df["Price"]
```

```python
output_file = r"D:\Analyst\DigitalProduct\Automated_Retail_Intelligence\data\processed\cleaned_online_data.csv"
df.to_csv(output_file,index=False)
```
